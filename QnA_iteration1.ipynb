```python

llm = HuggingFaceHub(repo_id="google/flan-t5-large", model_kwargs={"temperature":0, "max_length":100})
```

Loading Document


```python

loader =  UnstructuredPDFLoader("/content/drive/MyDrive/Data_PDFQ/AR_20050_TATAMOTORS_2021_2022_11062022223301.pdf")
documents = loader.load()
```

Having vector store first



```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=20)
```

```python
texts = text_splitter.split_documents(documents)
```

```python
!pip install sentence_transformers
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
docsearch = FAISS.from_documents(texts, embeddings)
```
